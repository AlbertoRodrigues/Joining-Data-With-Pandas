### Neste capítulo veremos alguns outros tipos de joins. Os joins apresentados serão:
1. Left Join
2. Right Join
3. Outer Join
4. Self Join

```python
import pandas as pd
import numpy as np
```

### O left join, right join e outer podem ser usados na função pd.merge com o argumento somente mudando o argumento para: "left", "right" ou "outer", respectivamente.

### O Left Join nada mais é do que uma junção das observações em que a chave dão match nas duas tabelas e das observações da tabela da esquerda que não dão match

```python
produtos=pd.DataFrame({"ID":np.arange(1,11), "tipo": ["A", "A", "A", "A", "A", "B", "B", "B", "B", "B"]})
produtos
```

```python
clientes=pd.DataFrame({"cliente": np.arange(1,8), "ID": [2, 2, 2, 3, 4, 7, 1]})
clientes
```

```python
pd.merge(produtos, clientes, on="ID", how="left")
```

### O Right Join nada mais é do que uma junção das observações em que a chave dão match nas duas tabelas e das observações da tabela da esquerda que não dão match

```python
clientes=pd.DataFrame({"cliente": np.arange(1,8), "ID": [2, 2, 2, 3, 4, 12, 101]})
clientes
```

```python
pd.merge(produtos, clientes, on="ID", how="right")
```

### O Outer Join nada mais é do que uma junção das observações em que a chave dão match nas duas tabelas e das observações tanto da tabela da esquerda tanto da tabela da direita que não dão match

```python
pd.merge(produtos, clientes, on="ID", how="outer")
```

## Outro tipo de join é self join, em que fazemos um join de duas tabelas iguais. Os joins podem ser inner, left, right, outer ou outro tipo de join, porém, necessariamente com a mesma tabela
